```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('FIFA.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
f, ax = plt.subplots(figsize=(10, 8))
x=df['Age']
sns.distplot(x,bins=10)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8,6))
x=df['Age']
ax = sns.distplot(x,hist=False,rug=True,bins=10)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8,6))
x=df['Age']
ax = sns.distplot(x,kde=False,rug=True,bins=10)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8,6))
ax = sns.countplot(x='Preferred Foot',data=df)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8,6))
ax = sns.stripplot(x='International Reputation',y='Potential',data=df)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8, 6))
ax = sns.violinplot(x=df["Potential"])
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.pointplot(x="International Reputation", y="Potential", hue="Preferred Foot", data=df, dodge=True)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.barplot(x="International Reputation", y="Potential", data=df)
plt.show()
```

```python
g = sns.relplot(x="Overall", y="Potential", data=df)
```

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(x="Height", y="Weight", data=df)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8, 6))
ax = sns.lineplot(x="Stamina", y="Strength", data=df)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8, 6))
ax = sns.regplot(x="Overall", y="Potential", data=df)
plt.show()
```
