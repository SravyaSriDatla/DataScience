```python
import openpyxl    #library to deal with excel sheets
import pandas as pd

workbook = openpyxl.Workbook()
sheet = workbook.active

data = [
    ['Name','Domain','Age','Location','Salary','Exp'],
    ['Alex','Testing',25,'Bangalore', 5000, 10],
    ['Barb','Java',30,'Chennai', 10000,3],
    ['Cherry','C',35,'Pune',15000,4],
    ['Deepan','DA',38,'Mumbai',20000,5],
    ['Sravya','DS',25,'Hyderabad',50000,8]
]

for row in data:
    sheet.append(row)

workbook.save('workshop_data.xlsx')

emp = pd.read_excel('workshop_data.xlsx')
print(emp)
```
